```python
# -- Monte Belo  T06

import pandas as pd
import matplotlib.pyplot as plt

# Dados das amostras
amostrasMb = [[375.8, 435.15, 389.35],
            [385.69, 449.74, 395.94], 
            [408.23, 473.52, 421.48], 
            [474.22, 416.30, 418.02], 
            [478.31, 407.69, 426.40], 
            [476.76, 408.71, 424.95]]


# Plotando o grfico 
df = pd.DataFrame(amostrasMb, columns=['MB1', 'MB2', 'MB3'], index=['07/04', '10/04','17/04', '19/04', '25/04', '28/04'])

df.plot(kind='line')
plt.title('Variao das amostras')
plt.xlabel('Dia')
plt.ylabel('Volume (cm)')
plt.grid()

plt.show()
print('------ Massa Saturada -------')
print('-----------------------------')
print(df)
```
